#### 01) loan_rate & loan_limit 결측값 확인

```python
def f_bp(data,col_name):
    bp_list = data['bank_product'].unique()
    
    for i in bp_list:
        bp = data[data['bank_product'] == i]
        print('bank_product : %s'%i)
        print('len_bp : ',len(bp[col_name]))
        print('%s 의 결측값'%i,bp[col_name].isnull().sum())
        print('q0 : %f \n q25 : %f \n q50 : %f \n q75 : %f \n q1 : %f \n '%(bp[col_name].quantile(q=(0)),bp[col_name].quantile(q=(0.25)),bp[col_name].quantile(q=(0.5)),bp[col_name].quantile(q=(0.75)),bp[col_name].quantile(q=(1))))
        sns.boxplot(data=bp, x=col_name,palette = 'Blues',orient='h')
        plt.title('bank_product Boxplot %s'%i)
        plt.show()
```

#### 02) loan_rate 전처리

+ bank_product 기준으로 평균 보간
+ 채워지지 않는 값은 application_id 기준으로 보간
    + 1.0 -> 가장 낮은 값으로 보간

```python
data['loan_rate'].isnull().sum()
```

```python
data['bank_product']
```

```python
ndf = []

for idx, val in data.groupby('bank_product'):
    #print(val)
    #print(idx)
    ndf.append(val.fillna(val['loan_rate'].mean()))
    
data_fill = pd.concat(ndf,axis=0, ignore_index=True)
```

```python
data_fill
```

```python
data_fill['loan_rate'].isnull().sum()
```

```python
line = []

for idx, val in data_fill.groupby('application_id'):
    #print(val)
    #print(idx)
    line.append(val.fillna(val['loan_rate'].min()))
                
data_fill_line = pd.concat(line,axis=0, ignore_index=True)
```
